## Part II: Exploring Spatial Data Components

Before we dive into answering questions about hospitals in California, let's first dive into what our spatial components mean.

### Polygons

Our first spatial component, polygons, are within our `geometry` column in our data. A **polygon** is the areas created from multiple data points are connected in a specific order to create an outline. We could think about county limits creating an outline that would determine a polygon of a county. 

Below we will look at the polygon that is created for all of the counties in California.

```python
import matplotlib.pyplot as plt

# this returns only 1 row per county or else our 
# county boundaries would be darker for counties
# with more hospitals
county_map = ca_hosp.drop_duplicates(subset = 'county')

# plotting only the geometry column to create polygons
county_map.plot(color = 'white',
             edgecolor = 'black',
             figsize = (8, 7))
plt.show()
```

### Coordinates

Another spatial component that is useful are coordinates. **Coordinates** are the are the building blocks for many spatial analyses. They are comprised of a pair of longitude (x-axis) and latitude (y-axis) values that when combined can be drawn as a point on a map.

Coordinates, or points, can be useful when we want to draw attention to contextual points on our map. 

Often, a combination of spatial components allows for maps to show multiple key data features. For instance, if we are interested in viewing every hospital's location in California, we can include each county as a polygon while using coordinates to plot every hospital to show where there are greater concentrations of hospitals in California.

<p style="color:red;">Requesting feedback: As shown in our `geometry` column, our polygons are automatically created since each polygon is made up of coordinates in a specific order to accurately draw each polygon. </p>

```python
county_map.plot(color = 'white',
            edgecolor = 'black',
            alpha = .25, # makes polygons more transparent
            figsize = (8, 7))
sns.scatterplot(x = 'longitude',
                y = 'latitude',
                s = 5, # size of the points
                data = ca_hosp)
plt.show()
```

### Centroids

Now that we have points for every hospital in California, we can incorporate another important spatial component. **Centroids** are the average values for longitude and latitude values. By getting averages for both longitudes and latitudes, we can then get a centroid coordinate. 

While we could calculate centroids for any set of coordinates, centroids are useful when we have a lot of coordinates and want a general understanding of where the center is for our points of interest. Using our points for hospitals in California, we will create and plot centroids for any county we want to explore.

```python
# county coordinates for our specific counties
county_of_interest = ca_hosp.loc[(ca_hosp['county'] == 'Nevada')]

county_of_interest.head()
```

```python
fig, ax = plt.subplots(figsize = (8, 7))

# plots the counties for the entirety of California
county_map.plot(ax = ax, # this allows us to have 2 polygons in layers
        color = 'white',
        alpha = .25,
        zorder = 1, # order of what plot to draw first TODO
        edgecolor = 'black')
# only plotting our county of interest's polygon
county_of_interest.plot(ax = ax,
                        color = 'gray',
                        alpha = .1, 
                        zorder = 2,
                        edgecolor = 'black')
# plotting the location of the hospitals in our county of interest
sns.scatterplot(x = 'longitude',
                y = 'latitude',
                color = 'maroon',
                s = 10, # size of the points
                data = county_of_interest)

# plots the hospitals in our specific county
plt.show()
```

Now that we can see where the hospitals are in our county of interest, let's see where our centroid would be located once we calculate the average longitude and latitude between our two counties. 

```python
import pandas as pd

hosp_centroid = pd.DataFrame({'longitude': county_of_interest['longitude'].mean(),
                              'latitude': county_of_interest['latitude'].mean()},
                             index = [0])
hosp_centroid
```

By getting the average between the hospitals in our county of interest, we now have the coordinates to plot the point of our centroid on our map of California.

This is also the reason why we need multiple data points to be able to calculate the average longitude and latitude. If we calculate the average longitude and latitude values for a county with only one hospital, we would get the same coordinates as our one hospital.

```python
cent_fig, cent_ax = plt.subplots(figsize = (8, 7))

county_map.plot(ax = cent_ax,
        color = 'white',
        alpha = .25,
        zorder = 1, # order of what plot to draw first TODO
        edgecolor = 'black')
# only plotting our county of interest's polygon
county_of_interest.plot(ax = cent_ax,
                        color = 'gray',
                        alpha = .1, 
                        zorder = 2,
                        edgecolor = 'white')
# plotting the actual coordinates for hospitals
sns.scatterplot(x = 'longitude',
                y = 'latitude',
                color = 'maroon',
                s = 10, # size of the points
                data = county_of_interest)
# plotting the hospital centroids we created for our specific counties
sns.scatterplot(x = 'longitude',
            y = 'latitude',
            s = 10,
            color = 'red',
            data = hosp_centroid)
plt.show()
```

### When to use: Polygons, Coordinates, or Centroids

In our hospital example, each spatial component is useful for different questions.

Polygons can be useful for when our data applies to the entire area. We will see more about this in the following applied section.

We can use coordinates when we want to see every data point. This could be helpful when we want to see where there are clusters of hospitals and where there are scarce numbers of hospitals.

Centroids could be helpful when there are too many data points to make sense of the clustering of data points. For instance, if we were looking at all hopsitals in the United States, it might be difficult to see trends when there are too many points. Instead, by using centroids, we could see the general center for where hospitals are in each state. 

Polygons = Area that is determined from an outline of points
Coordinates = 1 pair of longitude and latitude values
Centroids = Average coordinates for several points of interest

## Part I: Traditional Plots

Let's use traditional plots to look at life expectancy for states and the District of Columbia (DC).

```python
# horizontal bars makes reading state abbreviations easier
# change the x and y axes
bar = sns.barplot(x = 'life_expectancy',
            y = 'state',
            data = us_map)
bar.set(title = 'Average Life Expectancy Across the United States')

plt.show()
```

We can see slight differences between states, but it is difficult to see with so many different states near one another. Additionally, unless you are an expert at knowing where all 50 states and DC are located, it is hard to picture if neighboring states are similar.

For traditional plots, we might add additional components to better interpret state differences in life expectancy. One thing we could add is to change the color by region of the United States. Most often, states are broken down into four or five regions. Four regions tends to be the most common, which includes the West, the Northeast, the Midwest, and the South. 

Below we will try out incorporating region into our plot to helpfully make it more clear on trends in the data.

```python
region_bar = sns.barplot(x = 'life_expectancy', # '
            y = 'state',
            hue = 'region',
            data = us_map)
region_bar.set(title = 'Average Life Expectancy Across the United States')
# moves legend to the top right corner and outside of the plot
# makes this easier to see where the bars end
sns.move_legend(region_bar, 'upper left', bbox_to_anchor = (1, 1))

plt.show()
```

Now we can see there appears to be differences based on the region of the United States. However, the way the plot is organized right now makes it difficult to make comparisions between specific states.

This is a good example of a map being used to see differences between states. Static maps are good at visualizing trends when you have prior information, such as knowing where the states are at generally.

```python
us_map.plot(column = 'life_expectancy', # our life expectancy variable
            cmap = 'YlGnBu', # Colors range from yellow to green to blue
            edgecolor = 'black',
            legend = True)
plt.title('Life Expectancy in the United States')
plt.xlim([-180, -65]) # makes sure the entire US is included in the map
plt.show()
```

While this plot is helpful to start seeing how states are grouped together based on average life expectancy, we lose out on one component that was useful in our traditional methods, the actual values for each state.

In order to get the best of both plots, we can introduce another layer for more data exploration through the use of interactive plots.

```python
import plotly.express as px

state_map = px.choropleth(us_map,
                    geojson = us_map['geometry'], # draws our polygons
                    locations = us_map.index, # makes sure our polygons align with each state correctly
                    color = 'life_expectancy',
                    color_continuous_scale = px.colors.sequential.YlGnBu, # same color palette
                    hover_data = ['state_name', 'life_expectancy'], # what data is shown when hovering over a polygon
                    title = 'Life Expectancy in the United States (2025)',
                    scope = 'usa')
state_map.show()
```
